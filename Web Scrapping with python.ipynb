```python
#pip install pandas
#pis install bs4
import pandas as py
#pandas package use to conver the table format
from bs4 import BeautifulSoup
#BeautifulSoup is a function use to cover text file
import requests
#requests use to check the url permit or not permit
```

```python
permission=requests.get('https://www.91mobiles.com/list-of-phones/best-mobiles-under-100000-in-india')
permission
#if it return 200 is permit other wise not permit
```

```python
#BeautifulSoup use to conver or import the file
mob=BeautifulSoup(permission.text)
mob
```

```python
#use the google chrome add Selectorgadget and scrapping the 
Mob_name=mob.select('.name')
Mob_price=mob.select('.price_padding')
Mob_rate=mob.select('.finder_icon_storage_text')
Mob_ad=mob.select('.os_icon_cat')
Mob_bat=mob.select('.specs_li:nth-child(4) label:nth-child(1)')
```

```python
details=py.DataFrame(Mob_name,columns=['Mobile'])
details['Price']=Mob_price
details['Battery']=Mob_bat
details['Storage']=Mob_rate
details['Android_Version']=Mob_ad

details
```

```python
details.to_csv('C:/Users/Admin/Documents/DHARUN/Data Scrapping/Top_Mobiles.csv')
```
